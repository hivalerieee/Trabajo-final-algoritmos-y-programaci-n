```python
import os
import random
import pandas as pd

# Configuración inicial
NUM_ESTUDIANTES = 1000
CUPOS_POR_AULA = {
    1: 40, 2: 40, 3: 40, 4: 35, 5: 35,
    6: 35, 7: 25, 8: 25, 9: 25, 10: 10
}
PROMEDIO_MINIMO = 3.0
PORCENTAJE_APROBADOS = 0.7

# Malla curricular con nivel, asignatura, núcleo y créditos

nombres_cursos1 = ['Algebra y Trigonometría', 'Calculo Diferencial', 'Geometría Vectorial', 'Inglés I', 'Vivamos la U', 'Introducción a la Ingeniería']
semestre1 = [1, 1, 1, 1, 1, 1]
areas1 = ['Ciencias Básicas', 'Ciencias Básicas', 'Ciencias Básicas', 'Formación Complementaria', 'Formación Complementaria', 'Generales']
creditos1 = [3, 3, 3, 1, 1, 1]


nombres_cursos2 = ['Gestión de las Organizaciones', 'Habilidades Gerenciales', 'Algebra Lineal', 'Calculo Integral', 'Descubriendo la Física', 'Inglés II']
semestre2 = [2, 2, 2, 2, 2, 2]
areas2 = ['Administración y Finanzas', 'Administración y Finanzas', 'Ciencias Básicas', 'Ciencias Básicas', 'Ciencias Básicas', 'Formación Complementaria']
creditos2 = [3, 3, 3, 3, 3, 1]

nombres_cursos3 = ['Gestión Contable', 'Fisica Mecánica', 'Inglés III', 'Algoritmia y Programación', 'Probabilidad e Inferencia Estadística', 'Teoría General de Sistemas']
semestre3 = [3, 3, 3, 3, 3, 3]
areas3 = ['Administración y Finanzas', 'Ciencias Básicas', 'Formación Complementaria', 'Métodos Cuantitativos', 'Métodos Cuantitativos', 'Métodos Cuantitativos']
creditos3 = [3, 3, 1, 3, 3, 3]

nombres_cursos4 = ['Ingeniería Económica', 'Electiva en Física', 'Inglés IV', 'Diseño de Experimentos y Análisis de Regresión', 'Optimización', 'Gestión de Métodos y Tiempos']
semestre4 = [4, 4, 4, 4, 4, 4]
areas4 = ['Administración y Finanzas', 'Ciencias Básicas', 'Formación Complementaria', 'Métodos Cuantitativos', 'Métodos Cuantitativos', 'Producción, Logística y Calidad']
creditos4 = [3, 3, 1, 3, 3, 4]

nombres_cursos5 = ['Gestión Financiera', 'Laboratorio Integrado de Física', 'Inglés V', 'Formación Ciudadana y Constitucional', 'Dinámica de Sistemas', 'Muestreo y Series de Tiempo', 'Procesos Estocásticos y Análisis de Decisión', 'Gestión por Procesos']
semestre5 = [5,5,5,5,5,5,5,5]
areas5 = ['Administración y Finanzas', 'Ciencias Básicas', 'Formación Complementaria', 'Formación Complementaria', 'Métodos Cuantitativos', 'Métodos Cuantitativos', 'Métodos Cuantitativos', 'Producción, Logística y Calidad']
creditos5 = [3,1,1,1,3,3,3,3]

nombres_cursos6 = ['Gestión Tecnológica', 'Legislación', 'Electiva en Humanidades I', 'Inglés VI', 'Simulación Discreta', 'Formulación de Proyectos de Investigación', 'Normalización y Control de la Calidad']
semestre6 = [6,6,6,6,6,6,6]
areas6 = ['Administración y Finanzas', 'Administración y Finanzas', 'Electivas Socio Humanísticas', 'Formación Complementaria', 'Métodos Cuantitativos', 'Generales', 'Producción, Logística y Calidad']
creditos6 = [3,3,3,1,3,3,3]

nombres_cursos7 = ['Formulación y Evaluación de Proyectos de Inversión', 'Emprendimiento', 'Electiva en Humanidades II', 'Énfasis Profesional I', 'Electiva Complementaria I', 'Diseño de Sistemas Productivos']
semestre7 = [7,7,7,7,7,7]
areas7 = ['Administración y Finanzas', 'Administración y Finanzas', 'Electivas Socio Humanísticas', 'Énfasis Profesional', 'Formación Complementaria', 'Producción, Logística y Calidad']
creditos7 = [3,2,3,3,3,3]

nombres_cursos8 = ['Gestión de Proyectos', 'Electiva en Humanidades III', 'Énfasis Profesional I', 'Electiva Complementaria I', 'Administración de la Producción y del Servicio']
semestre8 = [8,8,8,8,8]
areas8 = ['Administración y Finanzas', 'Electivas Socio Humanísticas', 'Énfasis Profesional', 'Formación Complementaria', 'Producción, Logística y Calidad']
creditos8 = [3,3,3,3,3]

nombres_cursos9 = ['Electiva en Humanidades IV', 'Énfasis Profesional I', 'Electiva Complementaria I', 'Gestión de la Cadena de Abastecimiento', 'Ingeniería del Mejoramiento Continuo']
semestre9 = [9,9,9,9,9]
areas9 = ['Electivas Socio Humanísticas', 'Énfasis Profesional', 'Formación Complementaria', 'Producción, Logística y Calidad', 'Producción, Logística y Calidad']
creditos9 = [3,3,3,3,3]

nombres_cursos10 = ['Practica Profesional']
semestre10 = [10]
areas10 = ['Generales']
creditos10 = [12]


Semestre1 = {
    'Nombre': nombres_cursos1,
    'Semestre': semestre1,
    'Área': areas1,
    'Créditos': creditos1
}


Semestre2 = {
    'Nombre': nombres_cursos2,
    'Semestre': semestre2,
    'Área': areas2,
    'Créditos': creditos2
}

Semestre3 = {
    'Nombre': nombres_cursos3,
    'Semestre': semestre3,
    'Área': areas3,
    'Créditos': creditos3
}

Semestre4 = {
    'Nombre': nombres_cursos4,
    'Semestre': semestre4,
    'Área': areas4,
    'Créditos': creditos4
}

Semestre5 = {
    'Nombre': nombres_cursos5,
    'Semestre': semestre5,
    'Área': areas5,
    'Créditos': creditos5
}

Semestre6 = {
    'Nombre': nombres_cursos6,
    'Semestre': semestre6,
    'Área': areas6,
    'Créditos': creditos6
}

Semestre7 = {
    'Nombre': nombres_cursos7,
    'Semestre': semestre7,
    'Área': areas7,
    'Créditos': creditos7
}

Semestre8 = {
    'Nombre': nombres_cursos8,
    'Semestre': semestre8,
    'Área': areas8,
    'Créditos': creditos8
}

Semestre9 = {
    'Nombre': nombres_cursos9,
    'Semestre': semestre9,
    'Área': areas9,
    'Créditos': creditos9
}

Semestre10 = {
    'Nombre': nombres_cursos10,
    'Semestre': semestre10,
    'Área': areas10,
    'Créditos': creditos10
}


import pandas as pd

df1 = pd.DataFrame(Semestre1)

df2 = pd.DataFrame(Semestre2)

df3 = pd.DataFrame(Semestre3)

df4 = pd.DataFrame(Semestre4)

df5 = pd.DataFrame(Semestre5)

df6 = pd.DataFrame(Semestre6)

df7 = pd.DataFrame(Semestre7)

df8 = pd.DataFrame(Semestre8)

df9 = pd.DataFrame(Semestre9)

df10 = pd.DataFrame(Semestre10)

programa = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)

programa


# Función para generar código de asignatura
def generar_codigo_asignatura(row, consecutivo):
    nombre = row['Nombre']
    semestre = row['Semestre']
    creditos = row['Créditos']
    siglas = "".join([palabra[0].upper() for palabra in nombre.split()][:3])
    return f"{siglas}{semestre}{creditos}{consecutivo:02d}"

# Crear estructura de carpetas
def crear_estructura_carpetas():
    base_path = "Trabajo_Final"
    os.makedirs(base_path, exist_ok=True)

    for index, row in programa.iterrows():
        semestre = row['Semestre']
        asignatura = row['Nombre']
        semestre_path = os.path.join(base_path, f"Semestre_{semestre}")
        os.makedirs(semestre_path, exist_ok=True)
        asignatura_path = os.path.join(semestre_path, asignatura.replace(" ", "_"))
        os.makedirs(asignatura_path, exist_ok=True)

        excel_path = os.path.join(asignatura_path, "Archivo Excel")
        os.makedirs(excel_path, exist_ok=True)
        csv_path = os.path.join(asignatura_path, "Archivo CSV")
        os.makedirs(csv_path, exist_ok=True)


    return base_path

nombres_estudiantes_df = pd.read_csv('data.csv')
num_estudiantes = len(nombres_estudiantes_df)

def generar_notas_estudiantes(num_estudiantes, porcentaje_aprobados=0.7, promedio_minimo=3.0):

    notas = []
    num_aprobados = int(num_estudiantes * porcentaje_aprobados)

    # Generar notas para estudiantes aprobados
    for _ in range(num_aprobados):
        nota = round(random.uniform(promedio_minimo, 5), 2)
        notas.append(nota)

    # Generar notas para estudiantes reprobados
    for _ in range(num_estudiantes - num_aprobados):
        nota = round(random.uniform(1, promedio_minimo - 0.1), 2)  # -0.1 para evitar notas en el límite
        notas.append(nota)

    random.shuffle(notas)  # Mezclar las notas para una distribución aleatoria
    return notas

notas = generar_notas_estudiantes(num_estudiantes, porcentaje_aprobados=0.7, promedio_minimo=3.0)

# Agregar las notas al dataframe
nombres_estudiantes_df['Nota'] = notas
nombres_estudiantes_df['Semestre'] = 1  # Inicializar todos los estudiantes en el semestre 1

# Guardar el dataframe actualizado
from itertools import cycle
semestres = cycle(range(1, 11))
nombres_estudiantes_df['Semestre'] = [next(semestres) for _ in range(len(nombres_estudiantes_df))]


# Asignar estudiantes a grupos
def asignar_estudiantes_a_grupos(programa):
    estudiantes_df = pd.read_csv('estudiantes_con_notas.csv')
    resultados = []

    for index, row in programa.iterrows():
        semestre = row['Semestre']
        asignatura = row['Nombre']
        estudiantes_semestre = estudiantes_df[estudiantes_df['Semestre'] == semestre]

        if estudiantes_semestre.empty:
            continue

        # Salta si no hay estudiantes para este semestre

        cupo_aula = CUPOS_POR_AULA.get(semestre, 40)
        num_grupos = (len(estudiantes_semestre) + cupo_aula - 1) // cupo_aula
        estudiantes_asignados = 0

        for grupo in range(1, num_grupos + 1):
            estudiantes_en_grupo_df = estudiantes_semestre.iloc[estudiantes_asignados:estudiantes_asignados + cupo_aula]
            estudiantes_asignados += len(estudiantes_en_grupo_df)

            if estudiantes_en_grupo_df.empty:
                break

            promedio = estudiantes_en_grupo_df['Nota'].mean() if not estudiantes_en_grupo_df.empty else 0

            resultados.append({
                "Código Asignatura": generar_codigo_asignatura(row, grupo),
                "Nombre Asignatura": asignatura,
                "Número Total Estudiantes": len(estudiantes_en_grupo_df),
                "Código de Curso": grupo,
                "Notas": estudiantes_en_grupo_df['Nota'].tolist(),
                "Promedio": promedio
            })

    return resultados



base_path = crear_estructura_carpetas() # Llama a la funcion para crear carpetas
resultados = asignar_estudiantes_a_grupos(programa)

# Actualizar exportación de resultados para incluir columnas y formato correcto en Excel

def exportar_resultados_actualizado(resultados, base_path):
    estudiantes_df = pd.read_csv('estudiantes_con_notas.csv')
    for resultado in resultados:
        semestre = programa[programa['Nombre'] == resultado["Nombre Asignatura"]]['Semestre'].values[0] # Get semester from programa
        asignatura_path = os.path.join(base_path, f"Semestre_{semestre}", resultado["Nombre Asignatura"].replace(" ", "_"))
        os.makedirs(asignatura_path, exist_ok=True)

        archivo_base = f"{resultado['Código Asignatura']}-{resultado['Nombre Asignatura'].replace(' ', '')}-{len(resultado['Notas'])}-{resultado['Código de Curso']}"
        # Guardar en las subcarpetas de Excel y CSV, verificando si ya existen
        csv_file_path = os.path.join(asignatura_path, "Archivo CSV", f"{archivo_base}.csv")
        excel_file_path = os.path.join(asignatura_path, "Archivo Excel", f"{archivo_base}.xlsx")

        if os.path.exists(csv_file_path):
            os.remove(csv_file_path)
            print(f"El archivo CSV '{csv_file_path}' se eliminó.")

        if os.path.exists(excel_file_path):
            os.remove(excel_file_path)
            print(f"El archivo Excel '{excel_file_path}' se eliminó.")

        num_estudiantes_en_grupo = len(resultado["Notas"])
        nombres_estudiantes_df = pd.read_csv('data.csv')
        nombres_estudiantes = nombres_estudiantes_df['Nombre&Apellido'].head(num_estudiantes_en_grupo).tolist()

        data = pd.DataFrame({
            "Código Asignatura (CA)": [resultado["Código Asignatura"]] * len(resultado["Notas"]),
            "Número Total de Estudiantes (NTE)": [len(resultado["Notas"])] * len(resultado["Notas"]),
            "Código del Curso (CC)": [resultado["Código de Curso"]] * len(resultado["Notas"]),
            "Nombre Estudiante": nombres_estudiantes,
            "Nota Asignatura (NA)": resultado["Notas"]
        })
        data.to_csv(csv_file_path, index=False)
        data.to_excel(excel_file_path, index=False)

        data.to_csv(os.path.join(asignatura_path, "Archivo CSV", f"{archivo_base}.csv"), index=False)
        data.to_excel(os.path.join(asignatura_path, "Archivo Excel", f"{archivo_base}.xlsx"), index=False)

# Ejecutar de nuevo con la actualización
exportar_resultados_actualizado(resultados, base_path)

base_path


print("¡Proceso completado! Los datos se encuentran en la carpeta:", base_path)
```
